```python
import numpy as np
import seaborn as sns
import statistics
```

```python
df_data = sns.load_dataset('tips')
```

```python
df_data.head()
```

```python
print(df_data['total_bill'].mean())
print(df_data['total_bill'].median())
print(df_data['total_bill'].mode())
```

```python
statistics.mode(df_data['total_bill'])
```

```python
sns.boxplot(df_data['total_bill'])
```

```python
sns.histplot(df_data['total_bill'], kde=True) ## it is right skewed
```

```python
sns.histplot(np.log1p(df_data['total_bill']), kde=True) ## bit more normally distributed
```
